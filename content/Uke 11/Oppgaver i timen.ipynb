## Funksjoner i Python


Funksjoner er mye brukt i programmering. Du har nok allerede brukt funksjoner som:

```python
print()
int()
input()
```
Alle disse funksjonene er funksjoner som vi ikke har laget men som følger med som standard i Python.

Men hvorfor har vi funksjoner i det hele tatt?
Ser for deg at skriver kode som skal løse et problem. Og mens du løser problemet så skriver du mye kode som er veldig likt. Da hadde det vært greit å kunne slippe å skrive den samme koden om og om igjen. Det er her funksjoner kommer inn! Funksjoner som man skriver selv kan hjelpe med å skrive kode så man slipper å gjenta seg flere ganger. 

En funksjon er en samling med kodelinjer som blir kjørt når man bruker funksjonen. 
Funksjoner er en måte å kunne gjenbruke kode. Du kjører kode som du har skrevet fra før uten å måtte skrive alt på nytt. 

## Hvordan bruke funksjoner

Det å bruke en funksjon er veldig enkelt i Python. Du må vite navnet på funksjonen. Så kan du bruke/kjøre funksjonen eller som man sier på engelsk "call the function", med å bruke paranteser etter navnet på funksjonen.

Eksempel

```python
# Definerer en eksempelfunksjon
def sangteksten():
    print("Never gonna give you up")
    print("Never gonna let you down")
    print("Never gonna run around and desert you")

# Bruk av eksempelfunksjonen
sangteksten()

# Her bruker vi navnet på funksjonen, sangtekst
# Så skriver vi paranteser. Da "caller" man funksjonen og kjører linjene med kode i funksjonen
```

## Hvordan skrive funksjoner?

Funksjoner er vanlig å skrive selv når vi programmerer. I Python starter man det man kaller definisjonen på en funksjon med ordet 
```
def
```
```def``` er et spesielt ord (keyword) som sier at koden som vi skriver i neste kodeblokk er funksjonen. Etter ```def``` så skriver man navnet på funksjonen.
```
def navn_på_funksjon
```

Så skriver paranteser og kolon : ```def navn_på_funksjon():```. Man kaller det gjerne for funksjon definisjonen. All kode som skrives i neste kodeblokk kalles for funksjons kroppen. Denne kodeblokken eller funksjons kroppen er den koden som blir kjørt når man bruker funksjonen.

## Parametere / argumenter / innputt til funksjoner

Parametre eller argumenter eller input som det også kalles er variabler som blir satt til en verdi når man bruker funksjonen.

Ta den enkle matematiske funksjonen $f(x) = x^2$ skrevet som kode i Python:
```python
# Definere funksjonen
def f(x):
    return x**2
```
- Navnet på funksjonen er ```f```.
- Funksjonen tar inn bare ett argument med navn ```x```.
- ```x``` blir brukt som variabel i regnestykke ```x**2```
- Funksjonen gir tilbake resultatet av regnestykke ```x**2```

#### Bruk av funksjon med parametere
Når du kjører koden over vil du definere en funksjon som du kan bruke senere i koden/programmet.
Bruker du funksjonene som i koden som under:
```python
# Definere funksjonen
def f(x):
    return x**2

# Bruker funksjonen og lagrer det man får ut i variabelen svar
svar = f(5)
print(svar)
```
så skjer det et par ting:
1. Man definerer funksjonen som vist før.
2. Kaller funksjonen med tallet 5. Altså så blir alle steder ```x``` står blir til ```5```. Man gir ```x``` en verdi, nemlig verdien ```5```. Man sier gjerne at ```x``` argumentet fungerer som en (lokal) variabel i kodeblokken.

4. I uttrykket ```x**2``` blir x byttet ut med verdien 5: ```5**2```
5. Så returnerer funksjonen utregningen av ```5**2``` som er 25.
6. Variabelen svar blir satt til verdien av bruken av funksjonen hvor x er 5. Altså så er svar variabelen lik 25
7. Så printer man ut verdien til variabelen svar.
