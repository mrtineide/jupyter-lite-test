## Lister i Python

I Python og i andre programmeringsspråk trenger man ofte å lagre ting. Til nå har vi bare brukt variabler til å lagre data. Ting som man gjerne lager i variabler kan man også lagre flere av i en liste. Man bruker lister når man skal lagre masse data. Å lagre masse data med hver sin variabel hadde vært veldig mange variabler.


En liste er det man kaller en container. Akkuratt som en shipping container kan ha mange ting inni seg så kan en Python liste, som er av en container type, ha masse forskjellige data i seg. Lister i Python kan ha tall, streng med tekst, andre lister, variabler og mye annet.  

```python
from IPython.display import YouTubeVideo
display(YouTubeVideo('ohCDWZgNIU0'))
```

## For løkker i Python

Når man har en liste eller flere ting samlet i en container type, lister er container type, så er det veldig greit å gjøre en og en ting med hver ting i listen. Da er løkker det som gir mening å bruke.

Løkker i Python gjør at du kan kjøre den samme koden om igjen med hver enkelt "element" i listen. Man kan si at man _for_ hver _ting_ i _listen_ så kjør en kode. 



```python
from IPython.display import YouTubeVideo
display(YouTubeVideo('dHANJ4l6fwA'))
```
